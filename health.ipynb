```python
import requests

temp2 = []
fail_list = []
def geo(search):
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}'.format(search)
    headers = {
        "Authorization": "KakaoAK 5793cde5d6d57817d2643d29202e1b61"
    }
    places = requests.get(url, headers = headers).json()['documents']

    return (places[0]['x'], places[0]['y'])

for idx, _ in enumerate(temp):
    try : 
        temp2.append(geo(temp[idx]))
    except:
        fail_list.append(idx)


temp2
```
